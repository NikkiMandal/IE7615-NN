```python
df['gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
```

```python
fig = plt.figure(figsize=(10,8))
sns.countplot(x='gender', hue='race', data=df)
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.legend(title='Race', title_fontsize='10', loc='upper left')
plt.show()
```

```python
 numdf =df.select_dtypes(include=['float64','int64'])
 corr =numdf.corr()
 print(corr)
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(corr, annot=True,cmap='hot',fmt=".2f")
plt.xlabel("")
plt.ylabel("")
```
